```python
import pandas as pd
import matplotlib.pyplot as plt
result=pd.read_csv('./model_scores.csv')
```

```python
result
```

```python
# 假设 'result' 是已经准备好的数据
df = pd.DataFrame(result)

# 设置子图
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# 训练得分
bars = axes[0, 0].bar(df['Model'], df['Train Score'], color='blue')
axes[0, 0].set_title('Train Score Comparison')
axes[0, 0].set_ylabel('RMSE')
axes[0, 0].set_xticklabels(df['Model'], rotation=45)
# 在条形图上显示数值
for bar in bars:
    yval = bar.get_height()
    axes[0, 0].text(bar.get_x() + bar.get_width()/2, yval, round(yval, 4), ha='center', va='bottom', color='blue')

# 训练 R^2
bars = axes[0, 1].bar(df['Model'], df['Train R^2'], color='red')
axes[0, 1].set_title('Train R^2 Comparison')
axes[0, 1].set_ylabel('R^2')
axes[0, 1].set_xticklabels(df['Model'], rotation=45)
# 在条形图上显示数值
for bar in bars:
    yval = bar.get_height()
    axes[0, 1].text(bar.get_x() + bar.get_width()/2, yval, round(yval, 4), ha='center', va='bottom', color='red')

# 测试得分
bars = axes[1, 0].bar(df['Model'], df['Test Score'], color='green')
axes[1, 0].set_title('Test Score Comparison')
axes[1, 0].set_ylabel('RMSE')
axes[1, 0].set_xticklabels(df['Model'], rotation=45)
# 在条形图上显示数值
for bar in bars:
    yval = bar.get_height()
    axes[1, 0].text(bar.get_x() + bar.get_width()/2, yval, round(yval, 4), ha='center', va='bottom', color='green')

# 测试 R^2
bars = axes[1, 1].bar(df['Model'], df['Test R^2'], color='purple')
axes[1, 1].set_title('Test R^2 Comparison')
axes[1, 1].set_ylabel('R^2')
axes[1, 1].set_xticklabels(df['Model'], rotation=45)
# 在条形图上显示数值
for bar in bars:
    yval = bar.get_height()
    axes[1, 1].text(bar.get_x() + bar.get_width()/2, yval, round(yval, 4), ha='center', va='bottom', color='purple')

plt.tight_layout()
plt.show()
```
